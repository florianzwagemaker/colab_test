## Cleaning reads

This is just a basic example using FastP to clean some reads and write a new fastq file for downstream analysis

click the "play" button in the cell below and badabing badaboom the reads are cleaned

```python
%%bash

mkdir -p fastp_output

fastp --thread 1 \
    -i example_data/sars-cov-2_hong_kong_example.fastq.gz \
    -A -Q --cut_right \
    --cut_right_mean_quality 7 \
    --cut_right_window_size 20 \
    -l 100 -o  fastp_output/sars-cov-2_hong_kong_example_trimmed.fastq.gz \
    -h fastp_output/sars-cov-2_hong_kong_example_trimmed.html \
    -j fastp_output/sars-cov-2_hong_kong_example_trimmed.json
```
